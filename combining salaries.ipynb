```python
import pandas as pd
```

```python
espn = pd.read_csv('./datasets/espn_player_salaries_no_positions.csv')
dataworld = pd.read_csv('./datasets/dataworld_player_salaries.csv')
```

```python
dataworld_updated = dataworld[dataworld['season'] <= 2006][['name', 'team', 'salary', 'season']]
```

```python
combined_salaries = pd.concat([espn_updated, dataworld_updated], ignore_index=True)
```

```python
combined_salaries = combined_salaries.sort_values(by=['season', 'salary'], ascending=[True, False])
```

```python
future_salaries = pd.read_csv('./datasets/basketball_ref_player_salaries.csv')
future_salaries
```

```python
combined_salaries = pd.concat([combined_salaries, future_salaries], ignore_index=True)
```

```python
combined_salaries.dropna(axis=0, inplace=True)
combined_salaries
```

```python
combined_salaries.to_csv('./datasets/combined_player_salaries.csv', index=False)
```

```python
season_translate = pd.read_csv('seasons_translation.csv')
```

```python
combined_salaries = combined_salaries.rename(columns={'season': 'season_end'})
```

```python
combined_salaries_w_seasons = combined_salaries.merge(season_translate, on='season_end')
```

```python
combined_salaries_w_seasons[combined_salaries_w_seasons['team'].str.contains('New Orleans Hornets')]
```

```python
combined_salaries_w_seasons[combined_salaries_w_seasons['team'].str.contains('LA Clippers')]
```

```python
def replace_team_name(row):
    if row['team'] == 'NO/Oklahoma City Hornets':
        return 'New Orleans Hornets'
    if row['team'] == 'LA Clippers':
        return 'Los Angeles Clippers'
    return row['team']

combined_salaries_w_seasons.loc[:, ['team']] = combined_salaries_w_seasons.apply(replace_team_name, axis=1)
```

```python
combined_salaries_w_seasons[(combined_salaries_w_seasons['team'].str.contains('New Orleans Hornets')) & (combined_salaries_w_seasons['season_end'] == 2007)]
```

```python
combined_salaries_w_seasons[(combined_salaries_w_seasons['team'].str.contains('Los Angeles Clippers')) & (combined_salaries_w_seasons['season_end'] >= 2007)]
```

```python
teams_translation = pd.read_csv('teams_translation.csv')
```

```python
teams_translation
```

```python
combined_salaries_all = combined_salaries_w_seasons.merge(teams_translation, left_on='team', right_on='full_name')
combined_salaries_all
```

```python
combined_salaries_all = combined_salaries_all.drop('full_name', axis=1)
combined_salaries_all = combined_salaries_all.rename(columns={'abbreviation': 'team_abbr'})
combined_salaries_all
```

```python
combined_salaries_all.to_csv('./datasets/combined_player_salaries_extended.csv', index=False)
```
