```python
import pandas as pd
import numpy as np
```

```python
salaries_raw = pd.read_csv('./datasets/salaries_1985to2018.csv')
```

```python
salaries = salaries_raw[['player_id', 'salary', 'season_end', 'team']].copy()
```

```python
players_raw = pd.read_csv('./datasets/players.csv')
```

```python
players_raw.head(5)
```

```python
player_names = players_raw[['_id', 'name']].copy()
```

```python
salaries.merge(player_names, left_on='player_id', right_on='_id')
```

```python
player_salaries = salaries.merge(player_names, left_on='player_id', right_on='_id')
```

```python
player_salaries = player_salaries[['name', 'team', 'salary', 'season_end']]
```

```python
player_salaries
```

```python
player_salaries = player_salaries.rename(columns={'season_end': 'season'})
```

```python
player_salaries
```

```python
player_salaries = player_salaries.sort_values(by=['season', 'salary'], ascending=False)
```

```python
player_salaries.to_csv('dataworld_player_salaries.csv')
```
