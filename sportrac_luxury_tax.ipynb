```python
import requests as rq
import pandas as pd
from io import StringIO
from bs4 import BeautifulSoup
import random
import time
from IPython.display import clear_output
```

```python
url = f'https://www.spotrac.com/nba/cba/tax/'
site_request = rq.get(url).text
luxury_tax = pd.read_html(StringIO(str(site_request)), header=0)[0]
luxury_tax
```

```python
luxury_tax = luxury_tax.rename(columns={'Year': 'luxury_tax'})
```

```python
luxury_tax = luxury_tax.loc[:0, :]
```

```python
luxury_tax_melt = luxury_tax.melt(id_vars=['luxury_tax'], var_name='Season', value_name='Luxury Tax')
```

```python
luxury_tax_melt = luxury_tax_melt.dropna()
```

```python
luxury_tax_melt = luxury_tax_melt[['Season', 'Luxury Tax']]
```

```python
def season_end(row):
    return int(row['Season']) + 1

luxury_tax_melt.loc[:, 'Season'] = luxury_tax_melt.apply(season_end, axis=1)
luxury_tax_melt
```

```python
def text_to_number(row):
    return int(float(row['Luxury Tax'][1:-1]) * 1_000_000)

luxury_tax_melt.loc[:, 'Luxury Tax'] = luxury_tax_melt.apply(text_to_number, axis=1)
luxury_tax_melt
```

```python
luxury_tax_melt.to_csv('./datasets/luxury_tax.csv', index=False)
```
