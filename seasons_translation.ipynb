```python
import pandas as pd

seasons = [f'{year}-{str(year + 1)[-2:]}' for year in range(1996, 2024)]
seasons_end = [year + 1 for year in range(1996, 2024)]

season_data = {'season': seasons, 'season_end': seasons_end}

seasons_df = pd.DataFrame(season_data)
seasons_df
```

```python
seasons_df.to_csv('seasons_translation.csv', index=False)
```
