```python
url = f'https://www.spotrac.com/nba/cba'
site_request = rq.get(url).text
cap = pd.read_html(StringIO(str(site_request)))[0]
cap
```

```python
cap
```

```python
cap_years = cap[['Year', 'Cap Maximum']]
```

```python
def end_of_year(row):
    return row['Year'] + 1

cap_years.loc[:, 'Year'] = cap_years.apply(end_of_year, axis=1)
cap_years = cap_years.rename(columns={'Year': 'Season'})
cap_years
```

```python
cap_years.to_csv('./datasets/salary_cap.csv', index=False)
```
