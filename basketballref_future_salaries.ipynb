```python
import pandas as pd
from nba_api.stats.static import teams
```

```python
future_salaries_raw = pd.read_csv('future_nba_contracts.csv')
```

```python
future_salaries_raw
```

```python
future_salaries_raw.columns = future_salaries_raw.iloc[0]
future_salaries_raw = future_salaries_raw[1:]
future_salaries_raw
```

```python
future_salaries_raw = future_salaries_raw.drop(['-9999', 'Rk', 'Guaranteed'], axis=1)
future_salaries_raw
```

```python
future_salaries_raw = future_salaries_raw.rename(columns={'Player': 'name', 'Tm': 'team_abbr'})
future_salaries_raw
```

```python
future_salaries_melt = future_salaries_raw.melt(id_vars=['name', 'team_abbr'], var_name='season', value_name='salary')
future_salaries_melt
```

```python
future_salaries_melt = future_salaries_melt.dropna(axis=0)
future_salaries_melt
```

```python
def edit_season(row):
    return f'20{row["season"][-2:]}'
    
future_salaries_melt.loc[:, 'season'] = future_salaries_melt.apply(edit_season, axis=1)
future_salaries_melt.loc[:, 'season'] = future_salaries_melt['season'].astype(int)
future_salaries_melt
```

```python
future_salaries_melt.loc[:, 'salary'] = future_salaries_melt['salary'].str.replace('$', '')
future_salaries_melt.loc[:, 'salary'] = future_salaries_melt['salary'].astype(int)
future_salaries_melt
```

```python
future_salaries_melt
```

```python
future_salaries_melt = future_salaries_melt[future_salaries_melt['season'] > 2024]
future_salaries_melt
```

```python
combined_salaries = pd.read_csv('./datasets/combined_player_salaries.csv')
combined_salaries
```

```python
teams = pd.DataFrame(teams.get_teams())
```

```python
teams
```

```python
teams_translation = teams[['abbreviation', 'full_name']]
```

```python
future_salaries_melt = future_salaries_melt.merge(teams_translation, left_on='team_abbr', right_on='abbreviation')
```

```python
future_salaries_melt = future_salaries_melt[['name', 'full_name', 'salary', 'season']]
```

```python
future_salaries_melt = future_salaries_melt.rename(columns={'full_name': 'team'})
```

```python
future_salaries_melt = future_salaries_melt.sort_values(by=['season', 'salary'], ascending=[True, False])
```

```python
future_salaries_melt.to_csv('./datasets/basketball_ref_player_salaries.csv', index=False)
```
